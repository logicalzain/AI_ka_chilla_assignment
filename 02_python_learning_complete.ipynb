# While Loops

```python
# While loop
i = 0
while i < 3:
    print(i)
    i += 1
```

```python
# Break in while loop
i = 0
while i < 10:
    if i == 3:
        break
    print(i)
    i += 1
```

```python
# Continue in while loop
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

# Functions

```python
# Creating a function
def greet():
    print("Hello!")

greet()
```

```python
# Function with parameters
def greet(name):
    print(f"Hello, {name}!")

greet("zain")
```

```python
# Function with return
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
# Function with default arguments
def greet(name="World"):
    print(f"Hello, {name}!")

greet()
greet("zain")
```

```python
# Function with keyword arguments
def describe(name, age):
    print(f"{name} is {age} years old")

describe(age=25, name="zain")
```

# Error Handling

```python
# Try except block
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Try except with multiple exceptions
try:
    num = int("abc")
except ValueError:
    print("ValueError: Cannot convert to int")
except TypeError:
    print("TypeError")
```

```python
# Try except else finally
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print("Success:", result)
finally:
    print("Always executed")
```

### Generators

```python
# Example 1: Basic Generator
def my_generator():
    yield 1
    yield 2
    yield 3

for value in my_generator():
    print(value)
```

```python
# Example 2: Generator Expression
gen = (x**2 for x in range(5))
for value in gen:
    print(value)
```
