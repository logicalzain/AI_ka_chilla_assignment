
## Numpy operations

```python
# === NumPy Array Creation ===
arr1 = np.array([1, 2, 3, 4, 5])  # 1D array
arr2 = np.array([[1,2,3],[4,5,6]])  # 2D array
print("1D:", arr1)
print("2D:\n", arr2)

zeros = np.zeros((3,3))  # 3x3 zeros
ones = np.ones((2,4))  # 2x4 ones
identity = np.eye(3)  # 3x3 identity matrix
rng = np.arange(0, 20, 2)  # range with step
lin = np.linspace(0, 1, 5)  # 5 evenly spaced values
print("Zeros:\n", zeros)
print("Ones:\n", ones)
print("Identity:\n", identity)
print("Arange:", rng)
print("Linspace:", lin)
```

```python
# === NumPy Random ===
np.random.seed(42)
rand_uniform = np.random.rand(3, 3)  # uniform [0,1)
rand_normal = np.random.randn(3, 3)  # standard normal
rand_int = np.random.randint(1, 100, size=(3, 3))  # random integers
rand_choice = np.random.choice([10,20,30,40], size=5)  # random pick
print("Uniform:\n", rand_uniform)
print("Normal:\n", rand_normal)
print("Integers:\n", rand_int)
print("Choice:", rand_choice)
```

```python
# === NumPy Array Properties ===
arr = np.random.randint(1, 50, size=(4, 5))
print("Array:\n", arr)
print("Shape:", arr.shape)  # dimensions
print("Dtype:", arr.dtype)  # data type
print("Ndim:", arr.ndim)  # number of dimensions
print("Size:", arr.size)  # total elements
print("Itemsize:", arr.itemsize)  # bytes per element
```

```python
# === NumPy Indexing & Slicing ===
arr = np.arange(1, 26).reshape(5, 5)
print("Full Array:\n", arr)
print("Element [2,3]:", arr[2, 3])  # single element
print("Row 1:", arr[1])  # entire row
print("Col 2:", arr[:, 2])  # entire column
print("Subarray [1:3, 2:4]:\n", arr[1:3, 2:4])  # slice
print("Fancy indexing:", arr[[0,2,4], [1,3,0]])  # specific elements
print("Boolean mask (>15):", arr[arr > 15])  # conditional
```

```python
# === NumPy Math Operations ===
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])
print("Add:", a + b)  # element-wise addition
print("Sub:", a - b)
print("Mul:", a * b)
print("Div:", a / b)
print("Power:", a ** 2)
print("Sqrt:", np.sqrt(a))
print("Log:", np.log(a))
print("Exp:", np.exp(b))
print("Sin:", np.sin(a))
print("Dot product:", np.dot(a, b))  # dot product
```

```python
# === NumPy Aggregation ===
arr = np.random.randint(1, 100, size=(4, 5))
print("Array:\n", arr)
print("Sum:", arr.sum())  # total sum
print("Sum axis=0 (col):", arr.sum(axis=0))  # column-wise
print("Sum axis=1 (row):", arr.sum(axis=1))  # row-wise
print("Mean:", arr.mean())
print("Std:", arr.std())
print("Var:", arr.var())
print("Min:", arr.min(), "| Max:", arr.max())
print("Argmin:", arr.argmin(), "| Argmax:", arr.argmax())  # index of min/max
print("Cumsum:", arr[0].cumsum())  # cumulative sum
print("Cumprod:", np.array([1,2,3,4]).cumprod())  # cumulative product
```

```python
# === NumPy Reshape, Stack, Split ===
arr = np.arange(1, 13)
print("Reshape 3x4:\n", arr.reshape(3, 4))
print("Reshape 2x6:\n", arr.reshape(2, 6))
print("Flatten:", arr.reshape(3,4).flatten())  # back to 1D
print("Ravel:", arr.reshape(3,4).ravel())  # similar to flatten
print("Transpose:\n", arr.reshape(3,4).T)

x = np.array([1,2,3])
y = np.array([4,5,6])
print("Hstack:", np.hstack([x, y]))  # horizontal stack
print("Vstack:\n", np.vstack([x, y]))  # vertical stack
print("Column_stack:\n", np.column_stack([x, y]))
print("Split:", np.split(np.arange(9), 3))  # split into 3 parts
```

```python
# === NumPy Sorting, Unique, Where ===
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6])
print("Sorted:", np.sort(arr))
print("Argsort:", np.argsort(arr))  # indices that would sort
print("Unique:", np.unique(arr))  # unique values
print("Unique + counts:", np.unique(arr, return_counts=True))

# np.where = conditional selection
result = np.where(arr > 3, arr, 0)  # keep if >3, else 0
print("Where (>3):", result)

# clip values to range
print("Clip [2,5]:", np.clip(arr, 2, 5))
```


## PANDAS - DATA CLEANING

```python
# === Check Missing Values ===
print("=== Plot Data ===")
print(df.isnull().sum())  # missing count per column
print("Total missing:", df.isnull().sum().sum())
print("\n=== Titanic ===")
print(titanic.isnull().sum())
print("Missing %:\n", (titanic.isnull().sum() / len(titanic) * 100).round(2))
```

```python
# === Handle Missing Values - Titanic ===
titanic_clean = titanic.copy()  # work on copy

# fill numeric with median
titanic_clean['age'] = titanic_clean['age'].fillna(titanic_clean['age'].median())

# fill categorical with mode
titanic_clean['embarked'] = titanic_clean['embarked'].fillna(titanic_clean['embarked'].mode()[0])
titanic_clean['embark_town'] = titanic_clean['embark_town'].fillna(titanic_clean['embark_town'].mode()[0])

# drop column with too many nulls
titanic_clean = titanic_clean.drop(columns=['deck'])  # too many missing

print("After cleaning:")
print(titanic_clean.isnull().sum())
```

```python
# === Duplicates ===
print("Duplicates in df:", df.duplicated().sum())
print("Duplicates in titanic:", titanic.duplicated().sum())

# drop duplicates (if any)
df_clean = df.drop_duplicates()
print("Shape after dropping:", df_clean.shape)
```

```python
# === Data Type Conversion ===
df_clean = df.copy()
df_clean['Date_Listed'] = pd.to_datetime(df_clean['Date_Listed'])  # string to datetime
df_clean['Floor'] = df_clean['Floor'].astype(str)  # int to string (categorical)
print(df_clean.dtypes)
print("---")
print(df_clean['Date_Listed'].head())
```

```python
# === Rename Columns ===
df_renamed = df.rename(columns={'For Family': 'Family_Friendly', 'Area_SqFt': 'Area'})
print(df_renamed.columns.tolist())
```

```python
# === Replace Values ===
df_temp = df.copy()
df_temp['Furnished'] = df_temp['Furnished'].replace({'Yes': 1, 'No': 0})  # encode
df_temp['Terrace'] = df_temp['Terrace'].replace({'Yes': 1, 'No': 0})
df_temp['For Family'] = df_temp['For Family'].replace({'Yes': 1, 'No': 0})
df_temp.head()
```


## PANDAS - FEATURE ENGINEERING & TRANSFORMATIONS

```python
# === New Columns ===
df_eng = df.copy()
df_eng['Date_Listed'] = pd.to_datetime(df_eng['Date_Listed'])

df_eng['Price_Per_SqFt'] = df_eng['Price'] / df_eng['Area_SqFt']  # calculated column
df_eng['Rent_to_Price_Ratio'] = (df_eng['Rent'] / df_eng['Price'] * 100).round(2)  # ratio
df_eng['Month'] = df_eng['Date_Listed'].dt.month  # extract month
df_eng['Month_Name'] = df_eng['Date_Listed'].dt.month_name()  # month name
df_eng[['Location','Price','Area_SqFt','Price_Per_SqFt','Rent_to_Price_Ratio','Month_Name']].head(10)
```

```python
# === Binning / Cut ===
df_eng['Price_Category'] = pd.cut(df_eng['Price'], bins=3, labels=['Low','Medium','High'])
print(df_eng['Price_Category'].value_counts())
print("---")

# quantile-based binning
df_eng['Price_Quantile'] = pd.qcut(df_eng['Price'], q=4, labels=['Q1','Q2','Q3','Q4'])
print(df_eng['Price_Quantile'].value_counts())
```

```python
# === Apply, Map, Applymap ===
# apply - apply function to column/row
df_eng['Price_Log'] = df_eng['Price'].apply(np.log)  # log transform

# apply with lambda
df_eng['Price_Label'] = df_eng['Price'].apply(lambda x: 'Expensive' if x > 3000000 else 'Affordable')
print(df_eng['Price_Label'].value_counts())

# map - for Series
location_map = {'Haripur': 'Small City', 'Rawalpindi': 'Big City', 'Islamabad': 'Capital'}
df_eng['City_Type'] = df_eng['Location'].map(location_map)
print(df_eng[['Location', 'City_Type']].drop_duplicates())
```

```python
# === Titanic Feature Engineering ===
titanic_eng = titanic_clean.copy()
titanic_eng['family_size'] = titanic_eng['sibsp'] + titanic_eng['parch'] + 1  # total family
titanic_eng['is_alone'] = (titanic_eng['family_size'] == 1).astype(int)  # traveling alone
titanic_eng['age_group'] = pd.cut(titanic_eng['age'], bins=[0,12,18,35,60,100],
                                   labels=['Child','Teen','Adult','Middle','Senior'])
titanic_eng['fare_per_person'] = titanic_eng['fare'] / titanic_eng['family_size']
titanic_eng[['age','age_group','family_size','is_alone','fare_per_person']].head(10)
```


## PANDAS - MERGE, JOIN, CONCAT

```python
# === Concat - stack dataframes ===
df_haripur = df[df['Location'] == 'Haripur']
df_islamabad = df[df['Location'] == 'Islamabad']

combined = pd.concat([df_haripur, df_islamabad], ignore_index=True)  # vertical stack
print("Combined shape:", combined.shape)
print(combined['Location'].value_counts())
```

```python
# === Merge - like SQL join ===
# create two sample dataframes to merge
city_info = pd.DataFrame({
    'Location': ['Haripur', 'Rawalpindi', 'Islamabad'],
    'Population': [1000000, 3500000, 2200000],
    'Province': ['KPK', 'Punjab', 'Federal']
})

merged = df.merge(city_info, on='Location', how='left')  # left join on Location
print(merged[['Location','Price','Population','Province']].head(10))
```

```python
# === All merge types ===
left = pd.DataFrame({'key': ['A','B','C'], 'val1': [1,2,3]})
right = pd.DataFrame({'key': ['B','C','D'], 'val2': [4,5,6]})
print("Inner:\n", pd.merge(left, right, on='key', how='inner'))  # only matching
print("Left:\n", pd.merge(left, right, on='key', how='left'))  # all from left
print("Right:\n", pd.merge(left, right, on='key', how='right'))  # all from right
print("Outer:\n", pd.merge(left, right, on='key', how='outer'))  # all from both
```

```python
# === 1. Line Plot ===
fig, ax = plt.subplots(figsize=(10, 5))
for loc in df['Location'].unique():
    subset = df[df['Location'] == loc]
    ax.plot(subset.index, subset['Price'], marker='o', label=loc)
ax.set_title('Price Trend by Location')
ax.set_xlabel('Index')
ax.set_ylabel('Price')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# === 2. Bar Plot ===
avg_price = df.groupby('Location')['Price'].mean()
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# vertical bar
axes[0].bar(avg_price.index, avg_price.values, color=['#FF6B6B','#4ECDC4','#45B7D1'])
axes[0].set_title('Avg Price by Location (Vertical)')
axes[0].set_ylabel('Price')

# horizontal bar
axes[1].barh(avg_price.index, avg_price.values, color=['#FF6B6B','#4ECDC4','#45B7D1'])
axes[1].set_title('Avg Price by Location (Horizontal)')
axes[1].set_xlabel('Price')

plt.tight_layout()
plt.show()
```
