```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = sns.load_dataset('tips')
```

```python
sns.lmplot(x='total_bill', y='tip', data = df) #plot with best fit line
```

```python
x = df[['total_bill', "size"]]
y = df['tip']
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
model.predict([[35.527, 2]])
```

```python
df['smoker_num'] = df['smoker'].map({'Yes':1, 'No':0})
df.head()
```

```python
x = df[['total_bill', 'size', 'smoker_num']]
y = df['tip']
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
print(model.predict([[35.5, 2, 1]]))
print(model.predict([[35.5, 2, 0]]))
```

# classification with logistic regression

## binary classification in yes/no format

```python
from sklearn.linear_model import LogisticRegression
```

```python
x = df[['total_bill', "size", "tip"]]
y = df['smoker']
```

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
model.predict([[35.527, 2, 5.67]])
```

## naive bayes classification

```python
from sklearn.naive_bayes import GaussianNB as GNB
```

```python
x = df[['total_bill', "size", "tip"]]
y = df['smoker']
```

```python
m = GNB()
m.fit(x, y)
```

```python
m.predict([[35.527, 2, 5.67]])
```

## decision tree classification

```python
from sklearn.tree import DecisionTreeClassifier as DTC
```

```python
tree_model = DTC()
tree_model.fit(x, y)
```

```python
tree_model.predict([[35.527, 2, 5.67]])
```

## decidion tree regression
