```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
x = df[['total_bill']]
y = df['tip']
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
model.predict([[50]])
```
